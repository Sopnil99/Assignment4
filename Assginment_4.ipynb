**Feature transformation**

Feature transformation is a mathematical transformation in which we apply a mathematical formula to a particular column (feature) and transform the values, which are useful for our further analysis. It is a technique by which we can boost our model performance. It is also known as Feature Engineering, which creates new features from existing features that may help improve the model performance.

**Different Types of Distance Measures in Machine Learning**



1.   Geometric Distance Measures

There are different ways to measure how close things are to each other. Some ways use how far apart they are in space. There are different types of this way, like Minkowski, Euclidean, and Manhattan. Another way is called cosine similarity, which we will learn about later.

2.   Euclidean Distance
Euclidean Distance is a measure of the distance between two points in a two or multi-dimensional space. It is calculated as the square root of the sum of the squared differences between the corresponding coordinates of the two points. This distance metric is commonly used in machine learning and data science applications.

**Standardization**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# transform data
df3[['Marketing Spend','Administration','Transport']]=scaler.fit_transform(df3[['Marketing Spend','Administration','Transport']])
df3.head()
```

```python
for col in df3.columns:
    if df3[col].dtype == 'object':
        continue
    df6[col]= scaler.fit_transform(df3[[col]])
df3.head()  
```

**Log Transformation**

```python
from sklearn.preprocessing import FunctionTransformer
ft = FunctionTransformer(np.log1p)
df4[['Marketing Spend','Administration','Transport','Profit']]=ft.fit_transform(df4[['Marketing Spend','Administration','Transport','Profit']])
df4.head()
```

```python
for col in df4.columns:
    if df4[col].dtype == 'object':
        continue
    df4[col]= ft.fit_transform(df3[[col]])
df4.head()
```

**Max Absolute Scaler**

```python
from sklearn.preprocessing import MaxAbsScaler
max=MaxAbsScaler()
df5[['Marketing Spend','Administration','Transport']]=max.fit_transform(df5[['Marketing Spend','Administration','Transport']])
df5.head()
```

```python
for col in df5.columns:
    if df5[col].dtype == 'object':
        continue
    df5[col]= max.fit_transform(df5[[col]])
df5.head() 
```
